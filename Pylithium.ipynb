```python
# Remplacer les types de données en numériques

df_final = df_final.apply(pd.to_numeric)

df_final.info()
```

```python
# Classer par l'élément

df_final = df_final.sort_values(by = ['element', 'jour'])
df_final
```

```python
# Réindexer l'index + suppression de l'ancien index

df_final.reset_index(inplace=True)

df_final.drop('index', axis = 1, inplace = True)

df_final
```

```python
# Lissage de Savitzky-Golay des valeurs de la table

from scipy.signal import savgol_filter

df_final['tmv_lissee'] = 0
lissage = pd.DataFrame(columns=['a'])

for i in range(1, 112):
  pulse = df_final['tmv'][df_final['element']==i]
  tab_pulse = pulse.values

  liss_tmv = savgol_filter(tab_pulse, 21, 2)
   

  lissage = pd.concat([lissage,pd.DataFrame(liss_tmv)])

lissage.reset_index(inplace=True)

df_final['tmv_lissee'] = lissage.iloc[:,1]

df_final
```

```python
# Création des tests si TMV <2v et <2,5V

df_final['seuil2V'] = df_final['tmv_lissee'].apply(lambda x : 1 if x <=2 else 0 )
df_final['seuil2,5V'] = df_final['tmv_lissee'].apply(lambda x : 1 if x <=2.5 else 0 )

df_final
```

```python
# Calcul des coefficients directeurs entre chaque points d'un même élément

df_final['coeff_1pt'] = 0
df_final['calcul_jour_1pt'] = 0

df_final['coeff_6pts'] = 0
df_final['calcul_jour_6pts'] = 0

coef_mult = 10

for i in range(1, len(df_final)):
  if ((df_final.iloc[i-1,3] ) == (df_final.iloc[i,3] )) : # col 3 : element => test si l'élément est le même
      if df_final.iloc[i,0] != df_final.iloc[i-1,0] :  # col 0 : jour => test si le jour est le même
        df_final.iloc[i,7] = ((df_final.iloc[i,4]- df_final.iloc[i-1,4]) / (df_final.iloc[i,0] - (df_final.iloc[i-1,0]))) * coef_mult
        # col 7 : coeff_1pt - col 4 : tmv_lissee

        # Calcul du nombre de jours restants
        
        b = df_final.iloc[i,4] - ((df_final.iloc[i,7] * df_final.iloc[i,0]))
        df_final.iloc[i,8] = (-b) / df_final.iloc[i,7]
        
         

nbr_pts = 6        
for i in range(nbr_pts, len(df_final)):
  if ((df_final.iloc[i-nbr_pts,3] ) == (df_final.iloc[i,3] )) : # col 3 : element => test si l'élément est le même
      if df_final.iloc[i,0] != df_final.iloc[i-nbr_pts,0] :  # col 0 : jour => test si le jour est le même
        df_final.iloc[i,9] = ((df_final.iloc[i,4]- df_final.iloc[i-nbr_pts,4]) / (df_final.iloc[i,0] - (df_final.iloc[i-nbr_pts,0]))) * coef_mult
        # col 8 : coeff_6pts - col 4 : tmv_lissee

        # Calcul du nombre de jours restants
        b = df_final.iloc[i,4] - ((df_final.iloc[i,9] * df_final.iloc[i,0]))
        df_final.iloc[i,10] = (-b) / df_final.iloc[i,9]
        

df_final
```

```python
df_final['sum_coeff_1pt']= 0
for i in range(1, len(df_final)):
  if ((df_final.iloc[i-1,3] ) == (df_final.iloc[i,3] )) :
      df_final.iloc[i, 11] = (df_final.iloc[i-1,11] + df_final.iloc[i,7])
  else :
     df_final.iloc[i, 11] = df_final.iloc[i,7]

df_final['sum_coeff_6pts']= 0
for i in range(1, len(df_final)):
  if ((df_final.iloc[i-1,3] ) == (df_final.iloc[i,3] )) :
      df_final.iloc[i, 12] = (df_final.iloc[i-1,12] + df_final.iloc[i,9])
  else :
     df_final.iloc[i, 12] = df_final.iloc[i,9]


df_final
```

```python
# Lissage de la somme des coefficients directeurs

df_final['sum_coeff_lisse_1pt'] = 0
lissage = pd.DataFrame(columns=['a'])

for i in range(1, 112):
  pulse = df_final['sum_coeff_1pt'][df_final['element']==i]
  tab_pulse = pulse.values

  liss_tmv = savgol_filter(tab_pulse, 23, 2)
   

  lissage = pd.concat([lissage,pd.DataFrame(liss_tmv)])

lissage.reset_index(inplace=True)

df_final['sum_coeff_lisse_1pt'] = lissage.iloc[:,1]

df_final
```

```python
# Régression linéaire simple sur la somme des coeffs

from sklearn import model_selection
from sklearn.model_selection import cross_val_predict, cross_validate
from sklearn.linear_model import LinearRegression

elt = 3

temps = df_final[['jour']][(df_final['element']==elt)]
somme = df_final['sum_coeff_1pt'][df_final['element']==elt]

slr = LinearRegression()
slr.fit(temps, somme)

print(slr.intercept_)
print(slr.coef_)

pred_somme = slr.predict(temps)

plt.figure(figsize=(15,15))

plt.subplot(3,1,1)

plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['tmv'][(df_final['element']==elt)], label = 'TMV');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['tmv_lissee'][(df_final['element']==elt)], label = 'TMV_lissee');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['sum_coeff_1pt'][(df_final['element']==elt)], label = 'Somme coefs directeurs 1pt');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['sum_coeff_lisse_1pt'][(df_final['element']==elt)], label = 'Somme coefs directeurs 1pt');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['sum_coeff_6pts'][(df_final['element']==elt)], label = 'Somme coefs directeurs 6pts');
plt.plot(df_final['jour'][(df_final['element']==elt)], pred_somme, label = 'Pred somme coefs directeurs');
plt.xlabel("Temps (j)")
plt.ylabel("TMV (V)")
plt.title('TMV de l'+"'"+'élément '+ str(elt) +' à 20°C')
plt.legend()
plt.ylim(-4, 4) 

plt.subplot(3,1,2)
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['calcul_jour_1pt'][(df_final['element']==elt)], label = 'Jours restants 1pt');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['calcul_jour_6pts'][(df_final['element']==elt)], label = 'Jours restants 6 pts');


plt.legend()

plt.subplot(3,1,3)
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['coeff_1pt'][(df_final['element']==elt)], label = 'Coeff 1pt');
plt.plot(df_final['jour'][(df_final['element']==elt)], df_final['coeff_6pts'][(df_final['element']==elt)], label = 'Coeff 6 pts');
plt.legend()
```

```python
# Régression linéaire simple sur toute la table
df_final['reg_sum_coeff'] = 0
slr = LinearRegression()
prediction = pd.DataFrame(columns=['a'])

for i in range(1, 112):
  temps = df_final[['jour']][(df_final['element']==i)]
  somme = df_final['sum_coeff_1pt'][df_final['element']==i]
  slr = LinearRegression()
  slr.fit(temps, somme)
  pred = slr.predict(temps)

  test = pd.DataFrame(pred)
  prediction = pd.concat([prediction,test])


prediction.reset_index(inplace=True)

df_final['reg_sum_coeff'] = prediction.iloc[:,1]

df_final
```

```python
# Ajout d'une colonne quand la somme des coeffs lissés 1pt est inférieure ou égale à différentes valeurs

df_final['seuil_sum_coeff-2'] = df_final['sum_coeff_lisse_1pt'].apply(lambda x : 1 if x <= -2 else 0 )
df_final['seuil_sum_coeff-1'] = df_final['sum_coeff_lisse_1pt'].apply(lambda x : 1 if x <= -1 else 0 )

# Calcul de la somme des coefs max par élément et détermination d'un seuil de -1 par raport à ce max

df_final['sum_coeff_delta']=0
df_final['seuil_sum_coeff_delta']=0
df_final['seuil_sum_coeff_delta_0.5']=0
df_temp_final = pd.DataFrame(columns=['a'])

for i in range(1,112):
  df_temp1 = df_final['sum_coeff_lisse_1pt'][(df_final['element']==i)]
  maximum = df_temp1.max()
  df_temp2 = df_final['sum_coeff_delta'][(df_final['element']==i)]
  df_temp2 = df_temp1 - maximum

  df_temp_final = pd.concat([df_temp_final,df_temp2])


df_temp_final.reset_index(inplace=True)

df_final['sum_coeff_delta'] = df_temp_final.iloc[:,1]

df_final['seuil_sum_coeff_delta']=df_final['sum_coeff_delta'].apply(lambda x : 1 if x <= -1 else 0 )
df_final['seuil_sum_coeff_delta_0.5']=df_final['sum_coeff_delta'].apply(lambda x : 1 if x <= -0.5 else 0 )

df_final
```

```python
# Calcul du jour avant décroissance (dernier coeff directeur positif de la liste) et création du seuil

df_final['dern_jour'] = 0
df_temp_final = pd.DataFrame(columns=['a'])

for i in range(1,112):
  tmv_max = df_final['tmv_lissee'][(df_final['element']==i)].max() 
  jmax = df_final['jour'][(df_final['element']==i) & (df_final['tmv_lissee']==tmv_max)].max()
  
  df_temp = df_final['jour'][(df_final['element']==i)].apply(lambda x : 1 if x >= jmax else 0)

  df_temp_final = pd.concat([df_temp_final,df_temp])

df_temp_final.reset_index(inplace=True)

df_final['dern_jour'] = df_temp_final.iloc[:,1]

df_final
```
